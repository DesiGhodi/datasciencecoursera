```python
travelexp = pd.read_csv("why_share_travel_exp.csv")
travelexp #381 rows, 32 features
```

```python
structure = c.Structure()
structure.add_path(['IDEN'], ["PF", "AS", "COMP", "PJOY"])
structure.add_path(['INTER'], ["AM", "ER", "PR", "RR", "SR"])
structure.add_path(['PJOY'], ["PF", "AS", "COMP"])
structure.add_path(['COMP'], ["SR"])
structure.add_path(['AM'], ["ER", "PR", "RR", "SR"])
structure.add_path(['PF'], ["AS", "COMP"])
structure.add_path(['ER'], ["PR", "RR", "SR"])
structure.add_path(['AS'], ["COMP"])
structure.add_path(['PR'], ["RR", "SR"])
structure.add_path(['RR'], ["SR"])
```

```python
config = c.Config(structure.path(), scaled=False)
```

```python
config.add_lv_with_columns_named("IDEN", Mode.A, travelexp, "Iden")
config.add_lv_with_columns_named("INTER", Mode.A, travelexp, "Inter")
config.add_lv_with_columns_named("PJOY", Mode.A, travelexp, "Pjoy")
config.add_lv_with_columns_named("COMP", Mode.A, travelexp, "Comp")
config.add_lv_with_columns_named("AM", Mode.A, travelexp, "AM")
config.add_lv_with_columns_named("PF", Mode.A, travelexp, "PF")
config.add_lv_with_columns_named("ER", Mode.A, travelexp, "ER")
config.add_lv_with_columns_named("PR", Mode.A, travelexp, "PR")
config.add_lv_with_columns_named("RR", Mode.A, travelexp, "RR")
config.add_lv_with_columns_named("SR", Mode.A, travelexp, "SR")
```

```python
plspm_calc = Plspm(travelexp, config, Scheme.CENTROID)
```
